## Why Data Architecture Is Important

https://www.youtube.com/watch?v=CWHHnzWLQoU

## Summary
Without some form of architecture, data tends to reside in a state of disorder. All the cool cutting edge applications for data (prescriptive analytics, AI, advanced dashboarding) are not possible without some form a structure being applied to the data. That is our job, to architect structure to the data.

## Stakeholders

https://www.youtube.com/watch?v=2VCumrO3KAI

## Summary
Your stakeholders are the people or entities that will be relying on the data solution you design. Some examples of business stakeholders are:

Data Analysts: Their entire job is based around using and analyzing data. As an architect, you should try to understand their workflow and needs to design a dataflow that will help them in their jobs.

Clinicians: If you work in the medical field, clinicians like doctors, nurses, and registered dieticians need access to data to help care for patients. Knowing how these stakeholders use the data will help you in architecting a more user-friendly data environment.

Sales: Sales people need access to sales data quickly. While this is a gross over-generalization here, it is common to find sales staff that is not the most technically savvy. Working with these stakeholders to better understand their process will help you to architect a solution that they will more likely adopt.

It is best if you approach stakeholders from a business partnership point of view. This is no longer a support role. You will be working side by side with the business stakeholders to create and deliver a solution.

## When to Use or Not Use Data Architecture

https://www.youtube.com/watch?v=OBqEsU_PYgk

## Summary
Remember, as a data architect, your time and effort are important and need to be put to good use. Some red flags to look out for are if there is no specific or defined problem, there is no data, the problem does not align with the business, or it's a buzzword project. Learn to vet out projects where you can provide actual value.
